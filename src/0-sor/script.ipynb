#### 1. Realizar a importação dos módulos

```python
import requests
```

#### 2. Realizar a captura da data de início e fim de extração

```python
start_date_extraction = input("Informe a data de início da extração no formato yyyyMM (ex. 202501): ")
end_date_extraction = input("Informe a data de fim da extração no formato yyyyMM (ex.: 202512): ")

if start_date_extraction > end_date_extraction:
    raise Exception("A data de início da extração precisa ser obrigatóriamente menor que a data de fim da extração.")
```

#### 3. Realizar a construção e execução da função que cálcula as datas no formato "yyyy-MM"

```python
def years_and_months_between_two_dates(start_date_extraction : str, end_date_extraction : str) -> list:
    
    years_and_months_list = []
    
    while start_date_extraction <= end_date_extraction:

        if start_date_extraction[4:] <= "12":
            years_and_months_list.append(start_date_extraction[:4] + "-" + start_date_extraction[4:])
            start_date_extraction = str(int(start_date_extraction) + 1)
        else:
            start_date_extraction = str(int(start_date_extraction[:4]) + 1) + "01"

    return years_and_months_list


years_and_months_list = years_and_months_between_two_dates(start_date_extraction, end_date_extraction)
```

#### 4. Realizar a extração dos dados via requisição e armazenar os arquivos na camada SOR

```python
def extract_data_from_data_source(url_from_data_source, period) -> str:

    file_name = f"../../data/0-sor/tb-sor-fhvhv-trip-data-{period}.parquet"

    response = requests.get(url_from_data_source)

    if response.status_code == 200:        
        with open(file_name, "wb") as file:
            file.write(response.content)
        print(f"O arquivo {file_name} foi armazenado no diretório ./data/0-sor/ com sucesso...\n")
    else:
        print(f"Houve um erro {response.status_code} durante a captura do arquivo {file_name}...")


url_from_data_source = r"https://d37ci6vzurychx.cloudfront.net/trip-data/fhvhv_tripdata_{}.parquet"

for item in years_and_months_list:
    extract_data_from_data_source(url_from_data_source.format(item), item)
```
